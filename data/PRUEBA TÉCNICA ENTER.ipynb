## Findings from the cleaning and analysis for the dummies dataframe

The correlations seems to not match any of the possibilities to predict the fuel use and comsumption. We are going to try an unsupervised learning method to find any grouping possible.


```python
df_1.temp_inside[df_1.temp_inside== 0]
df_1=df_1.drop([93, 95, 97, 98, 99, 100, 102, 201, 203, 261, 267, 268], axis=0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
```

```python
clusterNum = 3
k_means = KMeans(init = "random", n_clusters = clusterNum, n_init = 10)
k_means.fit(df_1)
labels= k_means.labels_
df_1["Clus_km"] = labels
df_1.head(50)
```

```python
df_1.gas_type_converted.value_counts()
```

```python

fig, ax = plt.subplots(1,6, sharex=True );
fig.suptitle('Clustering distribution');
ax[0].set_title('cluster by distance');
ax[1].set_title('cluster by consume');
ax[2].set_title('cluster by speed');
ax[3].set_title('cluster by temp_inside');
ax[4].set_title('cluster by temp_outside');
ax[5].set_title('cluster by Gas type');
sns.histplot(ax= ax[0], data= df_1, y= 'distance', x= 'Clus_km');
sns.histplot(ax= ax[1], data= df_1, y= 'consume', x= 'Clus_km');
sns.histplot(ax= ax[2], data= df_1, y= 'speed', x= 'Clus_km');
sns.histplot(ax= ax[3], data= df_1, y= 'temp_inside', x= 'Clus_km');
sns.histplot(ax= ax[4], data= df_1, y= 'temp_outside', x= 'Clus_km');
sns.histplot(ax= ax[5], data= df_1, y= df_1.gas_type_converted, x= 'Clus_km');
```

```python
df_1.to_csv('clean_dataframe.csv')
```

```python
from sklearn.tree import DecisionTreeClassifier, plot_tree
```

```python
x= df_1.drop('Clus_km', axis=1)
y= df_1.Clus_km
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
tree= DecisionTreeClassifier(criterion="entropy")
tre1=tree.fit(x_train, y_train)
tree_pre= tre1.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix, r2_score, f1_score, classification_report
```

```python
r2_score(y_test, tree_pre)
```

```python
print(classification_report(y_test, tree_pre))
```
